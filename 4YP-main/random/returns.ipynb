```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file; parse the 'Date' column as datetime
csv_path = r"C:\Users\Sean\Documents\gml-master\data\straddle_features\features.csv"
df = pd.read_csv(csv_path, parse_dates=['Date'])

# Create an output directory named "plots" in the current working directory if it doesn't exist.
output_dir = os.path.join(os.getcwd(), 'plots')
os.makedirs(output_dir, exist_ok=True)

# Get all unique tickers from the dataframe
tickers = df['ticker'].unique()
# tickers = ["AAPL"]

# Loop through each ticker, filter the data, and plot the daily_returns over time
for ticker in tickers:
    df_ticker = df[df['ticker'] == ticker].sort_values(by='Date')
    
    plt.figure(figsize=(12, 6))
    plt.plot(df_ticker['Date'], df_ticker['daily_returns'], marker='o', linestyle='-')
    plt.title(f'Daily Returns for {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Daily Returns')
    plt.grid(True)
    plt.plot()
    # # Save the plot to the output directory with a filename based on the ticker.
    # output_file = os.path.join(output_dir, f'plot_{ticker}.png')
    # plt.savefig(output_file)
    # plt.close()  # Close the figure after saving to free memory
```
