# AAPL Delta Neutral Straddle Price Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load features data
df = pd.read_csv('4YP-main/data/straddle_features/features.csv')

# Filter for AAPL
aapl = df[df['ticker'] == 'AAPL'].copy()
aapl['Date'] = pd.to_datetime(aapl['Date'])
aapl = aapl.sort_values('Date')

print(f"AAPL records: {len(aapl)}")
print(f"Date range: {aapl['Date'].min()} to {aapl['Date'].max()}")
```

```python
# Plot straddle price over time
plt.figure(figsize=(14, 6))
plt.plot(aapl['Date'], aapl['straddle_price'], linewidth=0.8)
plt.xlabel('Date')
plt.ylabel('Straddle Price')
plt.title('AAPL Delta Neutral Straddle Price Over Time')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
